Question 1: What is a random variable in probability theory?
Answer: A random variable is a variable that stores the result of a random experiment as a number.

In simple words, when we do an experiment whose result is not fixed, the value we get is stored in a random variable.

Example:

 -Tossing a coin

  Head → 1

  Tail → 0

Here, the value (0 or 1) is a random variable.

Question 6: What is the Central Limit Theorem (CLT), and why is it critical in statistics?
Answer:- The Central Limit Theorem (CLT) says:
         If we take a large number of samples from any population, the distribution of the sample mean will become normal, even if the original data is not normal.

Why it is important:-
  i)Helps estimate population parameters
  ii)Allows use of normal distribution
  iii)Very useful in confidence intervals and hypothesis testing


Question 7: What is the significance of confidence intervals in statistical analysis?
Answer: A confidence interval gives a range of values where the true population value is likely to lie.

Example:
         If average salary = ₹30,000 ± ₹2,000
         Then true salary is likely between ₹28,000 and ₹32,000.
         It helps us understand how accurate our estimate is.

Question 8: What is the concept of expected value in a probability distribution?
Answer: The expected value is the average outcome of a random experiment over many trials.
It is calculated by: Expected Value=∑(value×probability)
Example:

If a game gives ₹100 with probability 0.5
Expected value = 100 × 0.5 = ₹50

```python
"""
Question 9: Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.

"""

import numpy as np
import matplotlib.pyplot as plt
data = np.random.normal(50, 5, 1000)

mean_value = np.mean(data)
std_value = np.std(data)

print("Mean:", mean_value)
print("Standard Deviation:", std_value)

plt.hist(data, bins=30)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Normal Distribution Histogram")
plt.show()
```

```python
"""
Question 10: You are working as a data analyst for a retail company. The company has
collected daily sales data for 2 years and wants you to identify the overall sales trend.
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
● Explain how you would apply the Central Limit Theorem to estimate the average sales
with a 95% confidence interval.
● Write the Python code to compute the mean sales and its confidence interval.
"""

"""
Answer:
(a) Applying Central Limit Theorem
    i)Take sample means from sales data
    ii)According to CLT, sample mean follows normal distribution
    iii)Use this to calculate 95% confidence interval
"""
import numpy as np
import scipy.stats as stats

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

mean_sales = np.mean(daily_sales)
std_sales = np.std(daily_sales, ddof=1)
n = len(daily_sales)

confidence_interval = stats.norm.interval(
    0.95,
    loc=mean_sales,
    scale=std_sales / np.sqrt(n)
)

print("Mean Sales:", mean_sales)
print("95% Confidence Interval:", confidence_interval)
```
